### If the dataset is not present then download from URL

```python
import os
from urllib.request import urlretrieve
url="https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2017-050118.txt"
local_fname="hurdat2.txt"

if not os.path.exists("hurdat2.txt"):
    urlretrieve(url,local_fname)
```

### Unprocessed dataset

```python
import pandas as pd

col_name=['date','time','record_id','status','latitude','longitude','max_wind','min_pressure','ne34ktr','se34ktr','sw34ktr','nw34ktr','ne50ktr','se50ktr','sw50ktr','nw50ktr','ne64ktr','se64ktr','sw64ktr','nw64ktr']

df=pd.read_csv("hurdat2.txt",header=None,names=col_name)
df


```

### Processed dataset

```python
cond_col=df['date'].str.get(0)
cond1=cond_col.str.isalpha()
df[['identifier','name','num_pts']]=df[cond1].dropna(axis='columns')
df2=df[['identifier','name','num_pts']].fillna(method='ffill')

cond2=cond_col.str.isdigit() # if the first character of date column is number then it is data
df=df[cond2]
df2[col_name]=df[df.columns[:20]]
df2=df2.dropna()
df=df2
df
```
